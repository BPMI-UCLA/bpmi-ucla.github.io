```python
root_dir = "images/impressions-covid/BPMISubmissions"
title_dir = "post-title.txt"
intro = "author-intro.txt"
feature_image = "feature-image.jpg"
submissions = "list.txt"
post_image = "covid-post-img.jpg"
post_file_path = "/Users/jackzhao/Downloads/BPMISubmissions/posts"
```

```python
template = f'''
---
title: "%(title)s"
date: 2021-02-15T18:00:00
featureImage: %(feature_image)s
postImage: {os.path.join(root_dir, post_image)}
draft: false
---

## About Author

%(author_intro)s

%(artwork_intro)s

%(artworks)s
'''

artwork = "!\[%(name)s\](%(path)s)"  
youtube_artwork = "{{< youtube %s >}}"
```

```python
os.chdir("/Users/jackzhao/Downloads/BPMISubmissions")

for project in os.listdir("."):
    
    artwork = "!\[%(name)s\](%(path)s)"  
    youtube_artwork = "{{< youtube %s >}}"

    if project.startswith(".") or os.path.isfile(project) or project == "posts":
        continue
    os.chdir(project)
    print(os.getcwd())
    

    works = []
    try:
        lists = open(submissions, "r")
        works = [x.strip() for x in lists.readlines()]
    except:
        works.append('youtube.txt')
    
    title = ''
    title_file = open(title_dir, "r")
    title = title_file.readline().strip()
    print (f" - Title: {title}")
    
    author_intro = ''
    try:
        author_file = open (intro, "r")
    except:
        author_file = open ("author-info.txt", "r")
    author_intro = ''.join(author_file.readlines())
    print(f" - Author:{author_intro[:20]}")
    
    artwork_intro = ''
    try:
        artwork_file = open(f"{project}.txt")
        artwork_intro = "## About Artwork\n" + ''.join(artwork_file.readlines())
    except:
        print(" - Oh Shit")
    
    artwork_strings = ""
    temp = []
    for artwork in works:
        if artwork == '':
            continue
        elif artwork == 'youtube.txt':
            f = open(artwork, 'r')
            tmp = f.readline().split('/')[-1]
            print('tmp: ', tmp)
            temp.append("{{< youtube %s >}}" % tmp)
            break
        elif '.txt' in artwork:
            content = open(artwork, "r")
            artwork_strings = ''.join(content.readlines())
            break
        else:
            n = artwork.split('.')[0]
            p = os.path.join(root_dir, project, artwork)
            temp.append("![" + n + "](../../" + p + ")")

    if temp != []:
        artwork_strings = '\n\n'.join(temp)

    
    feature_image = ""
    for file in os.listdir("."):
        if "feature" in file:
            feature_image = file
            break
    
    print(f"Creating...", os.path.join(post_file_path, project + ".md"))
    post_buffer = open(os.path.join(post_file_path, project + ".md"), "w")
    post_content = template % {
        'title':title,
        'author_intro': author_intro,
        'artwork_intro': artwork_intro,
        'artworks': artwork_strings,
        'feature_image': os.path.join(root_dir, project, feature_image)
    }
    print ('========') 
    print(post_content)
    
    print ('========') 
    post_buffer.write(post_content)
    post_buffer.close()
    os.chdir("..")
```
